```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.pipeline import make_pipeline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Assuming 'Work Interest', 'Mood Swings', 'Social Weakness', and 'Quarantine_Frustrations' columns exist in your DataFrame
features = df[['Work_Interest', 'Mood_Swings', 'Social_Weakness']].copy()
target = df['Quarantine_Frustrations']
```

```python
# Convert categorical variables to numeric using Label Encoding
label_encoder = LabelEncoder()
for column in features.select_dtypes(include='object').columns:
    features[column] = label_encoder.fit_transform(features[column])
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
# Use SMOTE to handle class imbalance
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
```

```python
# Use a pipeline with StandardScaler and RandomForestClassifier
rf_model = make_pipeline(StandardScaler(), RandomForestClassifier(random_state=42, class_weight='balanced'))
```

```python
# Hyperparameter tuning with specific values
param_grid_rf = {'randomforestclassifier__n_estimators': [100],
                 'randomforestclassifier__max_depth': [None],
                 'randomforestclassifier__min_samples_split': [2],
                 'randomforestclassifier__min_samples_leaf': [1]}
```

```python
# Fit the model
rf_model.set_params(randomforestclassifier__class_weight='balanced')  # Set class weight to 'balanced'
rf_model.fit(X_train_resampled, y_train_resampled)

# Evaluate RandomForest model
predictions_rf = rf_model.predict(X_test)
accuracy_rf = accuracy_score(y_test, predictions_rf)
conf_matrix_rf = confusion_matrix(y_test, predictions_rf)
classification_rep_rf = classification_report(y_test, predictions_rf, zero_division=1)
```

```python
# Display results for RandomForest
print("RandomForest Results:")
print(f'Accuracy: {accuracy_rf}')
print(f'Confusion Matrix:\n{conf_matrix_rf}')
print(f'Classification Report:\n{classification_rep_rf}')
```

```python
# Use a pipeline with StandardScaler and GradientBoostingClassifier
gb_model = make_pipeline(StandardScaler(), GradientBoostingClassifier(random_state=42))
```

```python
# Hyperparameter tuning with specific values for Gradient Boosting
param_grid_gb = {'gradientboostingclassifier__n_estimators': [100],
                 'gradientboostingclassifier__learning_rate': [0.1],
                 'gradientboostingclassifier__max_depth': [3]}
```

```python
# Fit the model
gb_model.fit(X_train_resampled, y_train_resampled)

# Evaluate Gradient Boosting model
predictions_gb = gb_model.predict(X_test)
accuracy_gb = accuracy_score(y_test, predictions_gb)
conf_matrix_gb = confusion_matrix(y_test, predictions_gb)
classification_rep_gb = classification_report(y_test, predictions_gb, zero_division=1)
```

```python
# Display results for Gradient Boosting
print("\nGradient Boosting Results:")
print(f'Accuracy: {accuracy_gb}')
print(f'Confusion Matrix:\n{conf_matrix_gb}')
print(f'Classification Report:\n{classification_rep_gb}')
```

```python
# Extract feature importances from the RandomForest model
feature_importances_rf = rf_model.named_steps['randomforestclassifier'].feature_importances_

# Extract feature importances from the Gradient Boosting model
feature_importances_gb = gb_model.named_steps['gradientboostingclassifier'].feature_importances_

# Display feature importances for RandomForest
feature_importance_df_rf = pd.DataFrame({'Feature': features.columns, 'Importance': feature_importances_rf})
feature_importance_df_rf = feature_importance_df_rf.sort_values(by='Importance', ascending=False)
```

```python
print("\nRandomForest Feature Importances:")
print(feature_importance_df_rf)

# Display feature importances for Gradient Boosting
feature_importance_df_gb = pd.DataFrame({'Feature': features.columns, 'Importance': feature_importances_gb})
feature_importance_df_gb = feature_importance_df_gb.sort_values(by='Importance', ascending=False)

print("\nGradient Boosting Feature Importances:")
print(feature_importance_df_gb)
```

```python
# Visualize feature importances
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
sns.barplot(x='Importance', y='Feature', data=feature_importance_df_rf, palette='viridis')
plt.title('RandomForest Feature Importances')

plt.subplot(1, 2, 2)
sns.barplot(x='Importance', y='Feature', data=feature_importance_df_gb, palette='viridis')
plt.title('Gradient Boosting Feature Importances')

plt.tight_layout()  # Add parentheses here
plt.show()
```
